```python
import pandas as pd
tabla_k= pd.read_csv("tabla_k.csv")
```

```python
from sklearn import preprocessing 
```

```python
# Se normalizan los datos con MinMax()
min_max_scaler = preprocessing.MinMaxScaler() 
tabla_k_escalado = min_max_scaler.fit_transform(tabla_k)
tabla_k_escalado = pd.DataFrame(tabla_k_escalado) 
tabla_k_escalado.head()
```

```python
# PASAR TODOS LOS VALORES A NÚMERICOS. 
tabla_k_escalado = tabla_k_escalado.rename(columns={
    0: 'basura_papel',
    1: 'basura_plastico',
    2: 'basura_resto',
    3: 'basura_organico',
    4: 'basura_cristal',
    5: 'edad',
    6: 'genero',
    7: 'estado_civil',
    8: 'hijos',
    9: 'tipo_vivienda',
    10: 'nivel_estudios',
    11: 'profesion',
    12: 'cant_aluminio',
    13: 'cant_plastico',
    14: 'cant_cristal',
    15: 'kg_papel',
    16: 'kg_organico'
})
```
