```python
# IMPORTAR BIBLIOTECA PANDAS PARA MANIPULACIÓN Y ANÁLISIS DATOS.
import pandas as pd
```

```python
# CREAR DATAFRAME DEL DOCUMENTO CSV. 
df = pd.read_csv('Dataset_CAT_sucio_archivo.csv')
df.head(1)
```

```python
# VERIFICAR SI HAY DATOS DUPLICADOS, EN ESTE CASO NO.
df.duplicated().sum()
```

```python
# SE ELIMINAN COLUMNAS NO DESEADAS.
columnas_a_eliminar = ["Codi municipi","Municipi","Comarca","Autocompostatge","Poda i jardineria","Residus voluminosos + fusta","RAEE","Ferralla","Olis vegetals","Tèxtil","Runes","Residus en Petites Quantitats (RPQ)", "Piles","Medicaments","Altres recollides selectives","Resta a Dipòsit","Resta a Incineració", "Resta a Tractament Mecànic Biològic","Resta (sense desglossar)",'R.S. / R.M. % total','F.R. / R.M. %']
df.drop(columnas_a_eliminar, axis=1, inplace=True)
```

```python
df.info()
```

```python
#DESPUÉS DE ELIMINAR COLUMNAS YA NO HAY NULOS.
df.isnull().sum()
```

```python
df.head(1)
```

```python
#SE AGRUPAN RESULTADOS POR AÑOS Y SE SUMAN TODOS LOS MUNICIPIOS PARA OBTENER DATOS DE CATALUÑA EN GENERAL.
sum_columnas = ['Població', 'Matèria orgànica', 'Paper i cartró', 'Vidre', 'Envasos lleugers', 'Total Recollida Selectiva', 'Suma Fracció Resta', 'Generació Residus Municipal Totals']
mean_columnas = ['Kg/hab/any recollida selectiva', 'Kg / hab / dia', 'Kg / hab / any']

df_sum = df.groupby(df['Any'])[sum_columnas].sum().round(2)
df_mean = df.groupby(df['Any'])[mean_columnas].mean().round(2)
df_resultado = pd.concat([df_sum, df_mean], axis=1)
```

```python
df_resultado.head(1)
```

```python
#SE AÑADEN DOS COLUMNAS NUEVAS.
df_resultado = df_resultado.assign(RS_RC_porcentaje_total=(df_resultado['Total Recollida Selectiva'] / df_resultado['Generació Residus Municipal Totals']) * 100).round(2)
df_resultado = df_resultado.assign(FR_RC_porcentaje_total=(df_resultado['Suma Fracció Resta'] / df_resultado['Generació Residus Municipal Totals']) * 100).round(2)
```

```python
df_resultado.head(1)
```

```python
# CAMBIAR NOMBRES DE COLUMNAS. 
df_resultado.rename(columns={'Població': 'Població Catalunya', 'Generació Residus Municipal Totals': 'Generació Residus CAT Totals','RS_RC_porcentaje_total':'RS / RC % Total','FR_RC_porcentaje_total':'FR / RC % Total','Suma Fracció Resta': 'Total Fracció Resta'}, inplace=True)
```

```python
df_resultado.head(1)
```

```python
#REORDENAR COLUMNAS.
df_resultado = df_resultado[['Població Catalunya', 'Matèria orgànica', 'Paper i cartró', 'Vidre','Envasos lleugers','Total Recollida Selectiva','Kg/hab/any recollida selectiva','Total Fracció Resta','Generació Residus CAT Totals','RS / RC % Total','FR / RC % Total','Kg / hab / dia','Kg / hab / any']]
```

```python
df_resultado.head(1)
```

```python
df_resultado.info()
```

```python
# CAMBIAR EL TIPO DE VARIABLE DE FLOAT A INT. 
df_resultado['Matèria orgànica'] = df_resultado['Matèria orgànica'].astype('int64')
df_resultado['Paper i cartró'] = df_resultado['Paper i cartró'].astype('int64')
df_resultado['Vidre'] = df_resultado['Vidre'].astype('int64')
df_resultado['Envasos lleugers'] = df_resultado['Envasos lleugers'].astype('int64')
df_resultado['Total Recollida Selectiva'] = df_resultado['Total Recollida Selectiva'].astype('int64')
df_resultado['Total Fracció Resta'] = df_resultado['Total Fracció Resta'].astype('int64')
df_resultado['Generació Residus CAT Totals'] = df_resultado['Generació Residus CAT Totals'].astype('int64')
```

```python
df_resultado.info()
```

```python
# SE ELIMINAN MÁS COLUMNAS PARA DEJAR MÁS LIMPIO EL DATASET.
df_resultado.drop(['Matèria orgànica','Paper i cartró','Vidre','Envasos lleugers', 'Kg/hab/any recollida selectiva', 'Kg / hab / dia', 'Kg / hab / any'],axis=1,inplace=True)
```
