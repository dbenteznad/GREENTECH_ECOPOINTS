```python
# IMPORTAR BIBLIOTECA PANDAS PARA MANIPULACIÓN Y ANÁLISIS DATOS.
import pandas as pd
```

```python
# CREAR DATAFRAME DEL DOCUMENTO CSV. 
df = pd.read_csv('Dataset_ESP_Sucio.csv')
df.head(2)
```

```python
# FILAS Y COLUMNAS
df.shape
```

```python
# ANALISIS GENERAL DEL DATASET.
df.describe()
```

```python
# ANALIZAR COLUMNAS Y TIPOS DE VARIABLE.
df.info()
```

```python
# IDENTIFICAR COLUMNAS DE MANERA ANALÍTICA.
df.columns
```

```python
# REORDENAR LAS FILAS ALREVÉS PARA EMPEZAR POR EL MENOR AÑO. 
df_reversed = df.iloc[::-1]
df_reversed.head()
```

```python
# CAMBIAR NOMBRES DE COLUMNAS. 
df_reversed.rename(columns={'RS / RC % Total': 'RS / RE % Total', 'FR / RC % Total': 'FR / RE % Total'}, inplace=True)
```

```python
df_reversed.head(1)
```

```python
# SE ELIMINAN COLUMNAS NO DESEADAS.
df_reversed.drop(['Reciclado','Compostage','Vertido','Incineración', 'RS / RE % Total','FR / RE % Total'], axis=1, inplace=True)
```

```python
df_reversed.head(1)
```

```python
#SE AÑADEN DOS COLUMNAS NUEVAS.
df_reversed = df_reversed.assign(RS_RE_porcentaje_total=(df_reversed['Total Recogida Selectiva'] / df_reversed['Total Generado España']) * 100).round(2)
df_reversed = df_reversed.assign(FR_RE_porcentaje_total= (df_reversed['Total Resto'] / df_reversed['Total Generado España']) * 100).round(2)
```

```python
df_reversed.head(1)
```

```python
# PARA CAMBIAR EL DTYPE HE BORRADO LAS DOS COLUMNAS TIPO OBJECT Y LAS HE CREADO DESDE JUPYTER PARA QUE SALIERA TIPO FLOAT.
df_reversed.info()
```

```python
# CAMBIAR NOMBRES COLUMNAS
df_reversed.rename(columns={'RS_RE_porcentaje_total': 'RS / RE % Total', 'FR_RE_porcentaje_total': 'FR / RE % Total'}, inplace=True)
```

```python
df_reversed.head(1)
```

```python
# GUARDAR DATAFRAME EN UN ARCHIVO CSV.
nombre_archivo = 'Dataset_ESP_Limpio.csv'
df_reversed.to_csv(nombre_archivo, index=False)
```
