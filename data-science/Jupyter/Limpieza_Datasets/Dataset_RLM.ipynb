```python
# ABRIMOS DE NUEVO EL DATAFRAME ACCIONES INICIAL PARA AGRUPAR LAS ACCIONES POR TIEMPO PARA FUTUROS ANÁLISIS DE DATOS.
```

```python
dft = pd.read_csv("acciones.csv")
```

```python
# CAMBIAR TIPO DE DATO OBJECT A DATETIME.
dft['fecha_accion'] = pd.to_datetime(dft['fecha_accion'])
dft['hora_accion'] = pd.to_datetime(dft['hora_accion'], format='%H:%M:%S').dt.time
```

```python
# Añadir la columna total_deposiciones (suma de las cantidades de todos los artículos).
dft['total_deposiciones'] = dft[['cant_aluminio', 'cant_plastico', 'cant_cristal', 'kg_papel', 'kg_organico']].sum(axis=1)
```

```python
dft.head()
```

```python
# Extraer el año y mes de la columna "realizado_en"
dft['ano_mes'] = dft['fecha_accion'].dt.to_period('M')

# Agrupar por ano_mes
grouped = dft.groupby(['ano_mes'])

# Calcular la suma de las columnas numéricas para cada grupo
summed = grouped[['cant_aluminio', 'cant_plastico', 'cant_cristal', 'kg_papel', 'kg_organico','total_deposiciones']].sum()

# Ordenar por ano_mes
dft_resumen = summed.sort_values(by=['ano_mes'])

# Mostrar el resultado
dft_resumen.head(12)
```

```python
# GUARDAMOS EL DATAFRAME DE ACCIONES POR TIEMPO. 
dft_resumen.to_csv("resumen_dft.csv", index=True)
```
