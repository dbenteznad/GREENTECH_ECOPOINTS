```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
```

```python
import pandas as pd
tabla_k= pd.read_csv("tabla_k_escalado.csv")
```

```python
tabla_k.head(1)
```

```python
# Carga el dataset de ejemplo
X = tabla_k

# Prueba diferentes valores de k
Nc = range(1, 11)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]
score
# Grafica los valores de k vs score
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
# Crear una instancia del modelo K-means
kmeans = KMeans(n_clusters=3, random_state=0)
# Ajustar el modelo a los datos
kmeans.fit(X)
# Obtener las etiquetas de los clústeres asignados a cada punto
labels = kmeans.labels_
# Obtener las coordenadas de los centroides
centroids = kmeans.cluster_centers_
# Visualización de los resultados
plt.scatter(X['genero'], X['kg_papel'], c=labels)
plt.scatter(centroids[:,0], centroids[:,5], marker='x', color='red')
plt.xlabel("Genero")
plt.ylabel("Kg_papel")
plt.title("K-means Clustering")
plt.show()
```
